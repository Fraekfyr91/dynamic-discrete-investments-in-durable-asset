```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
h_tau = pd.read_excel('htau.xlsx')
lbl = pd.read_excel('lbl.xlsx')
ccp2 = pd.read_excel('ccp_tau_2.xlsx')
ccp1 = pd.read_excel('ccp_tau_1.xlsx')
```

```python
ccp_tau1 = ccp1.values
ccp_tau2 = ccp2.values
```

```python
ccp_tau1.shape
```

```python
np.max(abs(ccp_tau1-ccp_tau2))
```

```python
choice1 = ccp_tau1.reshape(21,22,22)
choice2 = ccp_tau2.reshape(21,22,22)
#choice1.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm
# create a 3D array
X, Y = np.meshgrid(np.linspace(0, 21, 21), np.linspace(0, 22, 22))
Z1 = np.sum(choice1,1).T
Z2 = np.sum(choice2,2).T


# create a figure and an Axes3D object
fig = plt.figure()
ax = plt.axes(projection='3d')

# plot the 3D array as a surface
#ax.plot_surface(X, Y, Z1,  cmap=cm.hot)
ax.plot_surface(X, Y, Z2,  cmap=cm.pink)


# show the plot
plt.show()

```

```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
sns.set(style="whitegrid")
sns.set_palette("dark")
idx = np.arange(h_tau_vals.shape[0])
fig, ax = plt.subplots(figsize=(10,6))
tau = 1
bar1 = ax.bar(idx, h_tau_vals[:,tau]*10, label=lbl_types[tau][0])
tau -= 1
bar2 = ax.bar(idx, h_tau_vals[:,tau]*10, bottom=h_tau_vals[:,tau+1]*10, label=lbl_types[tau][0])

ax.set_xlim(-1,21)
ax.bar(52, h_tau_vals[-1,tau], bottom=h_tau_vals[-1,tau-1])

# Add gridlines
ax.grid(False, which='both', axis='x')
ax.set_axisbelow(True)

# Add title and axis labels
plt.xlabel('Housing Condition', fontsize=14, fontweight='bold')
plt.ylabel('Fraction of the Population', fontsize=14, fontweight='bold')

# Add legend
ax.legend()

# Remove box around the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Change fontsize for axis labels and tick marks
ax.tick_params(axis='both', labelsize=14)
ax.set_xticks(np.arange(h_tau_vals.shape[0]))
# Add values on top of bars
#for r1, r2 in zip(bar1, bar2):
#    h1 = r1.get_height()
#    h2 = r2.get_height()
#    ax.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%.2f" % (h1/10), ha="center", va="center", color="white", fontweight="bold", fontsize=14)
#    ax.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%.2f" % (h2/10), ha="center", va="center", color="white", fontweight="bold", fontsize=14)

# Save the figure as an image file
plt.savefig("holdings.png")
plt.show()
```
